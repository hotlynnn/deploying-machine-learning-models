```python
frog_data['ppt_mean'] = frog_data['ppt_mean'].fillna(frog_data['ppt_mean'].median())
frog_data['soil_mean'] = frog_data['soil_mean'].fillna(frog_data['soil_mean'].median())
frog_data['tmax_mean'] = frog_data['tmax_mean'].fillna(frog_data['tmax_mean'].median())
frog_data['tmin_mean'] = frog_data['tmin_mean'].fillna(frog_data['tmin_mean'].median())
```

```python
frog_data = frog_data[['eventDate', 'country', 'continent',
       'occurrenceStatus', 'ppt_mean', 'soil_mean', 'tmax_mean', 'tmin_mean',
       'Month']]
```
